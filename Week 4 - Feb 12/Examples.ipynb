### Concatenating Data Frames

```python
df4 = pd.DataFrame(data={'A':[1,2,3], 'B':[4,5,6]})
df4
```

```python
df5 = pd.DataFrame(data={'B':[10,11,12], 'C':[13,14,15]})
df5
```

```python
df6 = pd.concat([df4, df5], sort=True)
df6
```

```python
df7 = pd.DataFrame({'B':[7,8,9], 'C':[10,11,12], 'D':[13,14,15]})
df8 = pd.concat([df4, df7], axis=1)
df8
```
